```python
import pandas as pd
import numpy as np
```

Here we will load the data from a raw CSV file.

```python
df = pd.read_csv("comprehensive-data.csv")
df
```

I will replace the first row as the header:

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
```

```python
df.info()
```

Some of these columns are not needed.

```python
df = df.drop(df.columns[[2, 3,4,5,6,9,26]], axis=1)
```

```python
df.head(30)
```

Below, I will be deleting the games that have not been played yet and have NaN scores.

```python
df = df[df['Tm_Score'].notna()]
```

```python
df.head(30)
```

```python
df.info()
```

The OT, or overtime column is not needed.

```python
df = df.drop('OT', axis=1)
```

```python
df.info()
```

```python
df.to_csv('comprehensive-data.csv')
```
